```python
def getIndexesFromRanges(ranges: np.array):
    # get min and max indexes from array
    minRange = ranges[:, 0:1].flatten().min()
    maxRange = ranges[:, 1:2].flatten().max()
    indexes = np.empty([0], dtype=int)
    for i in range(minRange,maxRange+1):
        present = False
        for (f,t) in ranges:
            if f <= i and t >= i:
                present = True
        if (present):
            indexes = np.append(indexes, np.array([i]), axis=0)
    return indexes
```

```python
import json
import numpy as np

with open('squad_nbest_describe.json', 'r') as f:
    nbest_data = json.load(f)

with open('squad_tokens_describe.json', 'r') as f:
    token_data = json.load(f)

with open('squad_nbest.json', 'r') as f:
    nbest_orig = json.load(f)
    
# get indexes for each question
new_context = {}
nbest_data_filtered = ["5705eb3375f01819005e7768"]

for q in nbest_data:
    ranges = np.empty([0,2], dtype=int)
    question_result = nbest_data[q]
    tokens = token_data[q]
    nbest_orig_span = nbest_orig[q][0]
    startIndex = nbest_orig_span["start_index"]
    endIndex = nbest_orig_span["end_index"]
    ranges = np.append(ranges, [(startIndex,endIndex)], axis=0)
    for i in range(len(question_result)):
        probability = question_result[i]["probability"][0]
        start_logit = question_result[i]["start_logit"]
        end_logit = question_result[i]["end_logit"]
        startIndex = question_result[i]["start_index"]
        endIndex = question_result[i]["end_index"]
        if (start_logit + end_logit > -8):
            ranges = np.append(ranges, [(startIndex,endIndex)], axis=0)
    indexes = getIndexesFromRanges(ranges)
    new_tokens = np.empty([0])
    for i in indexes:
        new_tokens = np.append(new_tokens, tokens[str(i)])
    new_context[q] = " ".join(new_tokens)


new_context["5728e5224b864d1900165033"]
# 5728e5224b864d1900165033 
# 5728d6f02ca10214002da911 'Describe Hungarians' fails - better question wording? 
# 57107932a58dae1900cd69cc bad answer problem What was the name of the first Huguenot church in the New World? L'Église française à la Nouvelle-Amsterdam
# 5ad0228a77cf76001a686b58 palm springs problem (-4.379635810852051 -1.3898029327392578) Who visits Palm Springs for the beaches?
# 5705e4fe75f01819005e7704 to the east problem  (-4.32842493057251 -0.952679038047790 )
```

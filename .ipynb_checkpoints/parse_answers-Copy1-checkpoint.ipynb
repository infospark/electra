```python
import json
with open('squad_nbest.json', 'r') as f:
    nbest = json.load(f)
    
nbest_filtered = nbest #{"5705fd8475f01819005e7840": nbest["5705fd8475f01819005e7840"]}

desribe_candidates = {}

for qid in nbest_filtered:
    if qid in scores_0_5["gold"]:
        q = nbest_filtered[qid]
        matches = []
        for a in q:
            if a["start_logit"] + a["end_logit"] > -4:
                already_matched_range = False
                for m in matches:
                    if a["start_index"] >= m["start_index"] and a["start_index"] <= m["end_index"]:
                        already_matched_range = True
                    if a["end_index"] >= m["start_index"] and a["end_index"] <= m["end_index"]:
                        already_matched_range = True
                    if a["start_index"] <= m["start_index"] and a["end_index"] >= m["end_index"]:
                        already_matched_range = True
                # Seems that overlapping answers aren't all that useful to us        
                if already_matched_range == False:
                    matches.append(a)
        desribe_candidates[qid] = matches

desribe_candidates
```

```python
# compute token ranges for each question response
# consolidate ranges
# repeat until no more consolidation takes place

import numpy as np
def getIndexesFromRanges(ranges: np.array):
    # get min and max indexes from array
    minRange = ranges[:, 0:1].flatten().min()
    maxRange = ranges[:, 1:2].flatten().max()
    indexes = np.empty([0], dtype=int)
    for i in range(minRange,maxRange+1):
        present = False
        for (f,t) in ranges:
            if f <= i and t >= i:
                present = True
        if (present):
            indexes = np.append(indexes, np.array([i]), axis=0)
    return indexes
```
